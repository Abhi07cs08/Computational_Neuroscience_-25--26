```python
from pathlib import Path
import sys

def _ensure_repo_root_on_path(start: Path) -> Path:
    current = start.resolve()
    for candidate in [current, *current.parents]:
        if (candidate / "src").exists():
            repo_root = str(candidate)
            if repo_root not in sys.path:
                sys.path.insert(0, repo_root)
            return candidate
    raise RuntimeError("Could not locate project root containing 'src'")

REPO_ROOT = _ensure_repo_root_on_path(Path.cwd())
print(f"repo root on sys.path: {REPO_ROOT}")
```

```python
import optuna

db_path = REPO_ROOT / "optuna_studies" / "simclr_spec_reg_multi_hpo_study_no_var_warmup.db"
if not db_path.exists():
    raise FileNotFoundError(f"DB not found: {db_path}")

storage_url = f"sqlite:///{db_path.as_posix()}"
study = optuna.load_study(
    study_name="simclr_spec_reg_multi_hpo_study_no_var_warmup",
    storage=storage_url,
)

print("Using DB:", db_path)
print("Number of trials:", len(study.trials))
```
