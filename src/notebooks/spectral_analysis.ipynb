```python
from torch.utils.data import Dataset
from torchvision import transforms
class NeuralDataStimuli(Dataset):
    def __init__(self, benchmark_assembly, transforms=None):
        self.benchmark_assembly = benchmark_assembly
        self.stimulus_set = benchmark_assembly.stimulus_set
        self.transforms = transforms

    def __len__(self):
        return len(self.stimulus_set)

    def __getitem__(self, idx):
        stimulus_id = self.stimulus_set['stimulus_id'].values[idx]
        local_path = self.stimulus_set.get_stimulus(stimulus_id)
        img = image.imread(local_path)
        img = transforms.ToPILImage()(img)
        trsfm = transforms.Compose([
            transforms.Resize(32),
            transforms.CenterCrop(32),
            transforms.ToTensor(),
        ])
        if self.transforms:
            img = self.transforms(img)
        else:
            img = trsfm(img)
        neural_response = self.benchmark_assembly.data[idx]
        return img, neural_response
    
dataset = NeuralDataStimuli(benchmark_assembly)
print("dataset length:", len(dataset))
img, neural_response = dataset[0]
print("image shape:", img.shape)
print("neural response shape:", neural_response.shape)
```

```python
import torch

class BasicCNN(torch.nn.Module):
    def __init__(self, in_channels=1, out_channels=10):
        super(BasicCNN, self).__init__()
        self.conv1 = torch.nn.Conv2d(in_channels=in_channels, out_channels=16, kernel_size=3, stride=1, padding=1)
        self.pool = torch.nn.MaxPool2d(kernel_size=2, stride=2, padding=0)
        self.conv2 = torch.nn.Conv2d(in_channels=16, out_channels=32, kernel_size=3, stride=1, padding=1)
        self.fc1 = torch.nn.Linear(32 * 8 * 8, 128)
        self.fc2 = torch.nn.Linear(128, out_channels)

    def forward(self, x):
        x = self.pool(torch.relu(self.conv1(x)))
        x = self.pool(torch.relu(self.conv2(x)))
        print(x.shape)
        x = x.view(-1, 32 * 8 * 8)
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

```python
def fetch_activation_matrices(benchmark, model, activation_layer="fc1", transforms=None, num_iterations=None):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    print("using device:", device)
    model = model.to(device)
    benchmark_assembly = benchmark._assembly
    dataset = NeuralDataStimuli(benchmark_assembly, transforms=transforms)
    dataloader = torch.utils.data.DataLoader(dataset, batch_size=128, shuffle=False)
    all_neural_responses = []
    all_model_hidden_activations = []
    activations = {}
    def save_activation(name):
        def hook(module, inp, out):
            activations[name] = out.detach()
        return hook

    handles = []

    for name, module in model.named_modules():
        if name == activation_layer:
            handles.append(module.register_forward_hook(save_activation(name)))

    for i, (images, neural_responses) in enumerate(dataloader):
        if num_iterations is not None and i >= num_iterations:
            break
        images = images.to(device)
        neural_responses = neural_responses.to(device)
        _ = model(images)
        activations_batch = activations[activation_layer]
        all_neural_responses.append(neural_responses)
        all_model_hidden_activations.append(activations_batch)
    all_neural_responses = torch.cat(all_neural_responses, dim=0)
    all_model_hidden_activations = torch.cat(all_model_hidden_activations, dim=0)
    return all_neural_responses, all_model_hidden_activations
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import explained_variance_score

def ridge_regression(X, Y, alpha=1.0):
    # """Perform ridge regression to predict Y from X."""
    # n_features = X.shape[1]
    # I = torch.eye(n_features).to(X.device)
    # W = torch.linalg.inv(X.T @ X + alpha * I) @ X.T @ Y
    clf = Ridge(alpha=alpha)
    if torch.is_tensor(X):
        X = X.cpu().numpy()
    if torch.is_tensor(Y):
        Y = Y.cpu().numpy()
    clf.fit(X, Y)
    # W = torch.tensor(clf.coef_, device=X.device).T
    Y_pred = clf.predict(X)
    return Y_pred

def ev(Y_true, Y_pred):
    if torch.is_tensor(Y_true):
        Y_true = Y_true.cpu().numpy()
    if torch.is_tensor(Y_pred):
        Y_pred = Y_pred.cpu().numpy()
    return explained_variance_score(Y_true, Y_pred, multioutput='variance_weighted')
```

```python
def F_R_EV(benchmark, model, activation_layer="fc1", alpha=1.0, transforms=None, num_iterations=None):
    neural_responses, model_activations = fetch_activation_matrices(benchmark, model, activation_layer, transforms=transforms, num_iterations=num_iterations)
    neural_pred = ridge_regression(model_activations, neural_responses, alpha)
    F_EV = ev(neural_responses, neural_pred)

    model_pred = ridge_regression(neural_responses, model_activations, alpha)
    R_EV = ev(model_activations, model_pred)

    BPI = (2*F_EV*R_EV)/(F_EV + R_EV + 1e-10)

    scores = {'BPI': BPI, 'F_EV': F_EV, 'R_EV': R_EV}

    return scores
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np


def f(n, alpha, c, b):
    return c/((n)**alpha)

def fit_power_law(sorted_list_var_ratio, bounds=(10, 30)):
    i0, i1 = bounds
    if i1 > len(sorted_list_var_ratio):
        i1 = len(sorted_list_var_ratio)
    if bounds is not None:
        popt, pcov = curve_fit(f, range(i0, i1), sorted_list_var_ratio[i0:i1], p0=[0.01, 0, 0])
    else:
        popt, pcov = curve_fit(f, range(1, len(sorted_list_var_ratio)+1), sorted_list_var_ratio, p0=[0.01, 0])
    # popt, pcov = curve_fit(f, range(3, 15), sorted_list_var_ratio[3:15], p0=[0.01, 0])
    return popt, pcov
```

```python
def fetch_spectrum_pca(activation, n_components=40):
    activation = torch.tensor(activation).to(device)
    activation = activation.cpu()
    pca = PCA(n_components=n_components)
    X = activation.view(activation.size(0), -1)
    X_reduced = pca.fit_transform(X)
    sorted_list_var_ratio = sorted(pca.explained_variance_ratio_, reverse=True)
    sorted_list_var = sorted(pca.explained_variance_, reverse=True)
    return sorted_list_var_ratio, sorted_list_var
```

```python
import torch
def fetch_spectrum_svd(activation, n_components=40):
    activation = torch.tensor(activation).to(device)
    X = activation.view(activation.size(0), -1)
    X = X - X.mean(axis=0)
    with torch.no_grad():
        U, S, Vt = torch.linalg.svd(torch.tensor(X), full_matrices=False)
        eigvals = (S**2)
    eigvals = eigvals.cpu().numpy()
    eigvals = sorted(eigvals, reverse=True)
    eigvals = eigvals[:n_components]
    eigvals_normalized = eigvals / sum(eigvals)
    return eigvals_normalized,eigvals
```

```python
from sklearn.utils.extmath import randomized_svd

def fetch_spectrum_svd_randomized(activation, n_components=40):
    activation = torch.tensor(activation).to(device)
    X = activation.view(activation.size(0), -1)
    X = X - X.mean(axis=0)
    U, S, Vt = randomized_svd(X.cpu().numpy(), n_components=n_components, random_state=42, n_iter=2)
    eigvals = (S**2)
    eigvals = eigvals
    eigvals = sorted(eigvals, reverse=True)
    eigvals_normalized = eigvals / sum(eigvals)
    return eigvals_normalized,eigvals
```

```python
def fetch_spectrum_covar_eigh(activation, n_components=40):
    activation = torch.tensor(activation).to(device)
    X = activation.view(activation.size(0), -1)
    X = X - X.mean(axis=0)
    cov_matrix = X@X.T/(X.shape[0]-1)
    eigvals, _ = torch.linalg.eigh(cov_matrix)
    sorted_indices = np.argsort(eigvals.cpu().numpy())[::-1]
    sorted_eigvals = eigvals[sorted_indices]
    sorted_eigvals = sorted_eigvals[:n_components]
    sorted_list_var_ratio = sorted_eigvals / np.sum(sorted_eigvals)
    sorted_list_var = sorted_eigvals
    return sorted_list_var_ratio, sorted_list_var
```

```python
def fetch_spectrum_covar_eigen(activation, n_components=40):
    activation = torch.tensor(activation).to(device)
    X = activation.view(activation.size(0), -1)
    X = X - X.mean(axis=0)
    C = X @ X.T/(X.shape[0]-1)
    eigvals= torch.linalg.eigvalsh(C)
    eigvals = sorted(eigvals.cpu().numpy(), reverse=True)
    eigvals = eigvals[:n_components]
    eigvals_normalized = eigvals / sum(eigvals)
    return eigvals_normalized, eigvals
```

```python
def fetch_spectrum_covar_svd(activation, n_components=40):
    activation = torch.tensor(activation).to(device)
    X = activation.view(activation.size(0), -1)
    X = X - X.mean(axis=0)
    C = X @ X.T/(X.shape[0]-1)
    with torch.no_grad():
        U, S, Vt = torch.linalg.svd(C, full_matrices=False)
        # eigvals = (S**2)
        eigvals = S
    eigvals = eigvals.cpu().numpy()
    eigvals = sorted(eigvals, reverse=True)
    eigvals = eigvals[:n_components]
    eigvals_normalized = eigvals / sum(eigvals)
    return eigvals_normalized,eigvals
```

```python
def fetch_spectrum_covar_pca(activation, n_components=40):
    activation = torch.tensor(activation).to(device)
    activation = activation.cpu()
    X = activation.view(activation.size(0), -1)
    X = X - X.mean(axis=0)
    cov_matrix = X@X.T/(X.shape[0]-1)
    pca = PCA(n_components=n_components)
    pca.fit(cov_matrix)
    sqrt_eigvals = np.sqrt(pca.explained_variance_)
    sqrt_eigvals_ratio = sqrt_eigvals / np.sum(sqrt_eigvals)
    # sqrt_eigvals_ratio = np.sqrt(pca.explained_variance_ratio_)
    sorted_list_var_ratio = sorted(sqrt_eigvals_ratio, reverse=True)
    sorted_list_var = sorted(sqrt_eigvals, reverse=True)
    return sorted_list_var_ratio, sorted_list_var
```

```python
import torch

class BasicCNN(torch.nn.Module):
    def __init__(self, in_channels=1, out_channels=10):
        super(BasicCNN, self).__init__()
        self.conv1 = torch.nn.Conv2d(in_channels=in_channels, out_channels=16, kernel_size=3, stride=1, padding=1)
        self.pool = torch.nn.MaxPool2d(kernel_size=2, stride=2, padding=0)
        self.conv2 = torch.nn.Conv2d(in_channels=16, out_channels=32, kernel_size=3, stride=1, padding=1)
        self.fc1 = torch.nn.Linear(32 * 56 * 56, 128)
        self.fc2 = torch.nn.Linear(128, out_channels)

    def forward(self, x):
        x = self.pool(torch.relu(self.conv1(x)))
        x = self.pool(torch.relu(self.conv2(x)))
        x = x.view(-1, 32 * 56 * 56)
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

```python
import torch
from sklearn.decomposition import PCA
from scipy.optimize import curve_fit
import numpy as np


def f(n, alpha, c, b):
    # use numpy inside the model function (curve_fit will pass numpy arrays)
    n = np.asarray(n, dtype=float)
    return c / (n ** alpha) + b


def fit_power_law(sorted_list_var_ratio, bounds=(10, 30)):
    # robust wrapper around scipy.optimize.curve_fit
    arr = np.asarray(sorted_list_var_ratio, dtype=float)
    if arr.size == 0:
        return np.array([np.nan, np.nan, np.nan]), np.full((3, 3), np.nan)

    if bounds is not None:
        i0, i1 = bounds
        i0 = max(1, int(i0))  # ensure >=1 to avoid division by zero
        i1 = min(int(i1), arr.size)
        x = np.arange(i0, i1)  # same indexing as used elsewhere in the notebook
        y = arr[i0:i1]
    else:
        x = np.arange(1, arr.size + 1)
        y = arr

    if y.size < 3:
        # not enough points to fit 3 parameters
        return np.array([np.nan, np.nan, np.nan]), np.full((3, 3), np.nan)

    try:
        popt, pcov = curve_fit(f, x, y, p0=[0.5, 0.1, 0.0], maxfev=10000)
        # compute and print mean squared residuals for debugging (optional)
        residuals = y - f(x, *popt)
        msr = np.mean(residuals ** 2)
        # print(f"fit msr={msr:.6e}")
        return popt, pcov
    except Exception as e:
        print(f"fit_power_law: curve_fit failed: {e}")
        return np.array([np.nan, np.nan, np.nan]), np.full((3, 3), np.nan)


def fetch_spectrum_pca(activation, device=None, n_components=40):
    # keep backward compatibility: device is optional
    # accept either torch.Tensor or array-like
    if torch.is_tensor(activation):
        X_t = activation.detach().cpu()
    else:
        X_t = torch.tensor(activation).cpu()

    # flatten samples x features
    X = X_t.view(X_t.size(0), -1).numpy()

    # sklearn PCA expects 2D numpy
    pca = PCA(n_components=min(n_components, X.shape[1], X.shape[0]))
    pca.fit(X)
    # PCA already returns components sorted by explained variance
    var_ratio = np.array(pca.explained_variance_ratio_, dtype=float)
    var = np.array(pca.explained_variance_, dtype=float)
    # return full arrays (no further sorting necessary), but keep length n_components
    return var_ratio.tolist(), var.tolist()


# def spectral_loss(activation, device=None, n_components=40, target_alpha=1.0):
#     # compute spectrum, fit power law and return a scalar torch.Tensor on requested device
#     var_ratio, var = fetch_spectrum_pca(activation, device=device, n_components=n_components)
#     popt, pcov = fit_power_law(var_ratio, bounds=(10, 30))
#     alpha = popt[0]
#     # print(f"Fitted_alpha: {alpha}")
#     # fallback if fit failed
#     if np.isnan(alpha):
#         loss_value = 1.0
#     else:
#         loss_value = float(abs(alpha - target_alpha))
#     if device is None:
#         return torch.tensor(loss_value)
#     else:
#         # ensure returned tensor is on the same device as model tensors
#         return torch.tensor(loss_value, device=device)
```

```python
import torch

def spectral_loss(activation, device=None, n_components=40, target_alpha=1.0, bounds=(10, 30), eps=1e-12):
    # get spectrum (accepts torch or numpy; coerces to torch on chosen device)
    var_ratio, var = fetch_spectrum_pca(activation, device=device, n_components=n_components)
    if not torch.is_tensor(var_ratio):
        var_ratio = torch.as_tensor(var_ratio, device=device, dtype=torch.float32)
    else:
        if device is not None and var_ratio.device != torch.device(device):
            var_ratio = var_ratio.to(device)

    # fit power law slope on log–log with torch (no NumPy)
    lo, hi = bounds
    lo = max(1, int(lo))                           # skip the first component if desired
    hi = min(int(hi), var_ratio.shape[-1])
    if lo >= hi:                                   # simple fallback
        lo, hi = 1, max(2, var_ratio.shape[-1])

    y = torch.log(var_ratio[..., lo:hi] + eps)     # log power
    idx = torch.arange(lo, hi, device=var_ratio.device, dtype=var_ratio.dtype)
    x = torch.log(idx + 0.0)                       # log frequency index

    # normalize x to stabilize slope; do batch-agnostic cov/var
    x = (x - x.mean()) / (x.std() + eps)
    y = y - y.mean(dim=-1, keepdim=True)

    slope = (y * x).mean(dim=-1) / (x.pow(2).mean() + eps)
    alpha = -slope                                  # for k^{-alpha}, slope = -alpha

    # L1 distance to target (subgradient at 0 is well-defined in PyTorch)
    loss_value = (alpha - target_alpha).abs().mean()

    return loss_value if device is None else loss_value.to(device)
```

```python
def spectral_loss_diff(activation, device=None, n_components=40, target_alpha=1.0, bounds=(10, 30), eps=1e-12):
    X = activation.view(activation.size(0), -1)
    X = X - X.mean(axis=0)
    B, N = X.shape
    q=min(n_components, B, N)
    U, S, V = torch.pca_lowrank(X, q=q)
    eigvals = S**2
    eigvals = eigvals / (torch.sum(eigvals) + eps)
    min_idx = max(1, int(bounds[0]))
    max_idx = min(int(bounds[1]), eigvals.shape[-1])
    idx = torch.arange(min_idx, max_idx, device=eigvals.device, dtype=eigvals.dtype)
    y = torch.log(eigvals[min_idx:max_idx] + eps)
    x = torch.log(idx + 0.0)

    x = (x - x.mean()) / (x.std() + eps)
    y = y - y.mean(dim=-1, keepdim=True)
    slope = (y * x).mean(dim=-1) / (x.pow(2).mean() + eps)
    alpha = -slope
    loss_value = (alpha - target_alpha).abs().mean()
    return loss_value if device is None else loss_value.to(device)
```
